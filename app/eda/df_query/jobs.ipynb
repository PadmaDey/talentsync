```python
from rapidfuzz import process, fuzz

# Step 1: Define standard categories
standard_labels = ['Full-time', 'Not Specified', 'Not Provided', 'Part-time']

# Step 2: Get unique values from the column
raw_labels = df_klz['employment_type'].dropna().unique()

# Step 3: Build a fuzzy mapping dictionary
label_map = {}
for raw in raw_labels:
    match, score, _ = process.extractOne(raw, standard_labels, scorer=fuzz.token_sort_ratio)
    print(f"'{raw}' matched with '{match}' (score: {score})")  # Optional debug
    if score >= 70:  # You can lower threshold to 70 if needed
        label_map[raw] = match
    else:
        label_map[raw] = raw  # No close match, keep as is

# Step 4: Apply the mapping
df_klz['employment_type_normalized'] = df_klz['employment_type'].map(label_map)

# Step 5: Value counts
print(df_klz['employment_type_normalized'].value_counts())
```
