#### Now create a column with value of auth method count from the 'auth_methods' column list value

```python
import ast

df_extracted['auth_methods_count'] = df_extracted['auth_methods'].apply(
    lambda x: len(ast.literal_eval(x)) if isinstance(x, str) else (len(x) if isinstance(x, list) else 0)
)


df_extracted[['auth_methods','auth_methods_count']].head()
```

#### Filter the data based on newly created column 'auth_methods_count'

```python
df_extracted[df_extracted['auth_methods_count'] > 1]
```

#### Check the Fields which are included datatypes of 'object', 'string'
