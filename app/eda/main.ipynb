##### All the collections present in the database, dumping in the output directory in .csv format

```python
for coll_name in collection_names:
    collection = db[coll_name]
    documents = list(collection.find({}))

    if not documents:
        print(f"Skipped empty collection: {coll_name}")
        continue

    df = pd.json_normalize(documents)

    output_path = os.path.join(output_dir, f"{coll_name}.csv")
    df.to_csv(output_path, index=False)
    print(f"Dumped: {output_path}")

print("\nAll non-empty collections dumped successfully!")
```
